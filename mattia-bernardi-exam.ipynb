# PARTE 2: Trasformazione e Predizione

## 1.

```python
ds = pd.read_csv("data.csv")
ds.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, plot_confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
```

```python
#eventuale comando per ignorare i warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
X = ds.drop(columns='Bankrupt?')
Y = ds['Bankrupt?']
```

```python
Y.value_counts()
```

### Fortemente sbilanciato per quanto rigurda la classe da predirre

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,
                                                   train_size = 3/4,
                                                   test_size=1/4,
                                                   random_state=0,
                                                   stratify=ds['Bankrupt?'])
```

```python
modeldt = DecisionTreeClassifier()
modellr = LogisticRegression()
model_dummy = DummyClassifier(strategy = 'prior')

print("\nVALUTAZIONE MODELLO: DECISION TREE ")

modeldt.fit(X_train,Y_train)

print("\nRisultati su train")
predict = modeldt.predict(X_train)

plot_confusion_matrix(modeldt, X_train, Y_train, normalize='true')  
plt.show() 
print("\nAccuracy train:", accuracy_score(Y_train, predict),"\n")

print("\nRisultati su test")
predict = modeldt.predict(X_test)

plot_confusion_matrix(modeldt, X_test, Y_test, normalize='true')  
plt.show() 
print("\nAccuracy test:", accuracy_score(Y_test, predict),"\n")


print("\nVALUTAZIONE MODELLO: LOGISTIC REGRESSION ")

modellr.fit(X_train,Y_train)

print("\nRisultati su train")
predict = modellr.predict(X_train)

plot_confusion_matrix(modellr, X_train, Y_train, normalize='true')  
plt.show() 
print("\nAccuracy train:", accuracy_score(Y_train, predict),"\n")

print("\nRisultati su test")
predict = modellr.predict(X_test)

plot_confusion_matrix(modellr, X_test, Y_test, normalize='true')  
plt.show() 
print("\nAccuracy test:", accuracy_score(Y_test, predict),"\n")


print("\nVALUTAZIONE MODELLO: DUMMY CLASSIFIER ")

model_dummy.fit(X_train,Y_train)

print("\nRisultati su train")
predict = model_dummy.predict(X_train)

plot_confusion_matrix(model_dummy, X_train, Y_train, normalize='true')  
plt.show() 
print("\nAccuracy train:", accuracy_score(Y_train, predict),"\n")

print("\nRisultati su test")
predict = model_dummy.predict(X_test)

plot_confusion_matrix(model_dummy, X_test, Y_test, normalize='true')  
plt.show() 
print("\nAccuracy test:", accuracy_score(Y_test, predict),"\n")
```

### Considerazioni:

* In tutti i modelli abbiamo ottenuto una accuracy elevata sia su train che su test, questo è dovuto fortemente dal fatto che anche sbagliando tutti gli esempi della classe 1, essendo in numero estremamente inferiore rispetto alla classe 0, l'accuracy non subisce un drop eccessivo.

* Come possiamo vedere dalle confusion matrices, la classe 1 infatti non viene praticamente mai predetta correttamente.

* Il modello che si comporta in maniera meno peggiore su questo aspetto è il DT


## 2.

```python
num_fold = 5

modeldt = DecisionTreeClassifier()
modellr = LogisticRegression()
model_dummy = DummyClassifier(strategy = 'prior')

print("\nVALUTAZIONE MODELLO: DECISION TREE ")

modeldt.fit(X_train,Y_train)

cv_results = cross_val_score(modeldt, X, Y, cv=num_fold, scoring='accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_results.mean(), cv_results.std() * 2))


print("\nVALUTAZIONE MODELLO: LOGISTIC REGRESSION ")

modellr.fit(X_train,Y_train)

cv_results = cross_val_score(modellr, X, Y, cv=num_fold, scoring='accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_results.mean(), cv_results.std() * 2))


print("\nVALUTAZIONE MODELLO: DUMMY CLASSIFIER ")

model_dummy.fit(X_train,Y_train)

cv_results = cross_val_score(model_dummy, X, Y, cv=num_fold, scoring='accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_results.mean(), cv_results.std() * 2))
```

### Le accuratezze sono rimaste pressoché invariate.

## 3.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, chi2
```

```python
model_lr = LogisticRegression()

pp = Pipeline(steps=[
    ('scaler', MaxAbsScaler()),
    ('nomr', Normalizer())
])

pp_pca = Pipeline(steps=[
    ('pp', pp),
    ('pca', PCA(n_components=10)),
    ('model', model_lr)
])

pp_skb = Pipeline(steps=[
    ('pp', pp),
    ('skb', SelectKBest(chi2)),
    ('model', model_lr)
])

print("\nVALUTAZIONE PCA + LR")

pp_pca.fit(X_train,Y_train)

cv_results = cross_val_score(pp_pca, X, Y, cv=num_fold, scoring='accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_results.mean(), cv_results.std() * 2))

plot_confusion_matrix(pp_pca, X_test, Y_test, normalize='true')  
plt.show()

print("\n\nVALUTAZIONE SKBest + LR")

pp_skb.fit(X_train,Y_train)

cv_results = cross_val_score(pp_skb, X, Y, cv=num_fold, scoring='accuracy')
print("Accuracy: %0.2f (+/- %0.2f)" % (cv_results.mean(), cv_results.std() * 2))

plot_confusion_matrix(pp_skb, X_test, Y_test, normalize='true')  
plt.show() 
```

### Il modello sia con PCA che con SKBest presenta la medesima accuratezza ma sbaglia sempre a predirre la classe 1

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters_pca = {
    'pp__nomr__norm': ['l1', 'l2', 'max'],
    'pca__n_components': [5, 7, 10, 12],
    'model__penalty': ['l1', 'l2', 'elasticnet'],
    'model__solver': ['newton-cg', 'lbfgs', 'liblinear']
}

gs_clf = GridSearchCV(pp_pca, parameters_pca,  cv=5, n_jobs=-1)
gs_clf.fit(X_train, Y_train)

gs_clf.best_params_

pp_pca.set_params(**gs_clf.best_params_)
# Preprocessing of training data, fit model 
pp_pca.fit(X_train, Y_train)

# Preprocessing of validation data, get predictions
preds = pp_pca.predict(X_test)

# Evaluate the model on test
score = accuracy_score(Y_test, preds)
print('Accuracy Score on test:', score)
plot_confusion_matrix(pp_pca, X_train, Y_train, normalize='true')
plt.show()
```
